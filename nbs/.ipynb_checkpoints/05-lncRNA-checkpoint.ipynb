```python
pwd
```

```python
#clean up fasta ID to just be contig id
#MIGHT not be needed
!cut -d' ' -f1 {fasta} \
> {wd}transcriptome.fa
```

```python
!head {wd}transcriptome.fa
```

```python
!fgrep ">" {wd}transcriptome.fa 
```

```python
!cat {wd}transcriptome.fa | perl -e '$count=0; $len=0; while(<>) {s/\r?\n//; s/\t/ /g; if (s/^>//) { if ($. != 1) {print "\n"} s/ |$/\t/; $count++; $_ .= "\t";} else {s/ //g; $len += length($_)} print $_;} print "\n"; warn "\nConverted $count FASTA records in $. lines to tabular format\nTotal sequence length: $len\n\n";' \
| sort > {wd}transcriptome.tab
```
